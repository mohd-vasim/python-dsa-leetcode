```python
# 1. Sum of list of elements

numbers = [1, 2, 3, 4, 5]
print(sum(numbers))
```

```python
# Find largest element in a list

numbers = [3, 8, 2, 10, 5]

# Assume you have largest element
largest_elem = numbers[0]

for num in numbers[1:]:
    if num > largest_elem:
        largest_elem = num

print(largest_elem)
```

```python
# 3. Remove a duplicate from a list

numbers = [3, 8, 2, 2, 10, 5]
new_list = []

for num in numbers:
    if num not in new_list:
        new_list.append(num)
    else:
        print(f"Duplicate: {num}")


print(new_list)
```

```python
# 4. Check if all elements in a list are unique

numbers = [3, 8, 2, 2, 10, 5]
unique_elems = []

for num in numbers:
    if num not in unique_elems:
        unique_elems.append(num)
    
unique_elems
```

```python
# 5. Reverse a list

numbers = [3, 8, 2, 10, 5]
reversed_list = []

for i in range(1, len(numbers)+1):
    reversed_list.append(numbers[-i])

print(reversed_list)
```

```python
# 6. Count number of Odd and Even elements in a list

numbers = [3, 8, 2, 10, 5]
odds = []
evens = []

for num in numbers:
    if num % 2 == 0:
        odds.append(num)
    else:
        evens .append(num)

print(f"There are {len(evens)} even numbers: {evens}")
print(f"There are {len(odds)} odd numbers: {odds}")
```

```python
# 7. Check if a list is subset of another list

lst1 = [1, 2, 3]
lst2 = [1, 2, 3, 4, 5]

sublist = []

for num in lst1:
    if num in lst2:
        sublist.append(num)


print(sublist)
print(True if sublist else False)
```
