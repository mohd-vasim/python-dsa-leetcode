```python
# 3. Remove a duplicate from a list

numbers = [3, 8, 2, 2, 10, 5]
new_list = []

for num in numbers:
    if num not in new_list:
        new_list.append(num)
    else:
        print(f"Duplicate: {num}")


print(new_list)
```

```python
# 8. Maximum current_diff between consecutive elements in a list

lst = [1, 7, 3, 10, 5]

def check_max_consecutive(lst: list):
    """Check max consecutive"""

    # Edge case
    if len(lst) < 2:
        return 0

    max_diff = 0

    for i in range(1, len(lst)):

        current_diff = abs(lst[i] - lst[i - 1])

        if current_diff > max_diff:

            max_diff = current_diff

    return max_diff

print(check_max_consecutive(lst))

print(check_max_consecutive([10, 11, 15, 3]))
```

This logic pattern is called **"Two-Pointer" or "Sliding Window"** approach, more specifically:

- **Two-Pointer Technique**: You're comparing two adjacent elements (`lst[i]` and `lst[i-1]`) as you iterate through the list
- **Linear/Sequential Traversal**: Moving through the list one step at a time, comparing consecutive pairs

In this case, it's examining **adjacent elements** to find the maximum difference between them. This is a common pattern used for:

- Comparing consecutive elements
- Finding gaps or differences between nearby items
- Processing paired data points

**Other names for similar patterns:**
- **Pairwise comparison**: Comparing each pair of consecutive elements
- **Window of size 2**: A minimal sliding window that only looks at two elements at a time

This approach is efficient with **O(n) time complexity** since you iterate through the list once, making only one pass with constant operations per iteration.

```python
# 9. Merge two sorted list

list1 = [1, 3, 5]
list2 = [2, 4, 6]

# Options:
# 1. merge two lists and use sorted function
# 2. 

def merge_two_sorted_lists(list1: list, list2: list):
    """Merge two sorted lists"""

    # Two pointers
    i, j = 0, 0
    new_list = []

    # Traverse until i or j as same as lists length
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            new_list.append(list1[i])

            # Update i for next comparison
            i += 1

        else:
            new_list.append(list2[j])

            # Update j for next comparison
            j += 1
        
    # If lists have different length
    while i < len(list1):
        new_list.append(list1[i])
        i += 1

    while j < len(list2):
        new_list.append(list2[j])
        j += 1
    
    return new_list

print(merge_two_sorted_lists(list1, list2))
```

```python
def rotate_list(lst, k):
    """Rotate a list"""

    if not lst:
        return []

    new_list = []
    n = len(lst)
    k %= n   # fix 1

    i = n - k

    while i < n:
        new_list.append(lst[i])
        i += 1

    i = 0
    while i < n - k:
        new_list.append(lst[i])
        i += 1

    return new_list

print(rotate_list(lst=[1, 2, 3, 4, 5], k=2))
print(rotate_list(lst=[10, 20, 30, 40, 50], k=3))
```

```python
# Another approach

def rotate_list(lst, k):
    n = len(lst)
    if n == 0:
        return []
    
    k %= n
    res = [0]*n
    
    for i in range(n):
        res[(i+k) % n] = lst[i]
    
    return res

print(rotate_list(lst=[1, 2, 3, 4, 5], k=2))
print(rotate_list(lst=[10, 20, 30, 40, 50], k=3))
```
